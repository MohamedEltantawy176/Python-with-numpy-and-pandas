### Real-World Examples Using Lists in Python

Lists are one of the most commonly used data structures in Python, thanks to their versatility and ease of use.
Here are several practical examples that illustrate their use in real-world scenarios


### Example 1: Manage A To Do List

* Create a To Do List To Keep Track OF Tasks

```python
to_do_list = ["Buy groceries", "Clean the house", "Pay bills"]

# Add a new task
to_do_list.append("Walk the dog")
to_do_list.append("Read a book")

# Remove a completed task
to_do_list.remove("Pay bills")

# check if a task is in the list
if "Clean the house" in to_do_list:
    print("Don't forget to clean the house")

# Print the current to do list
print("Current To Do List:")
for task in to_do_list:
    print("-", task)
```

### Example 2: Organizing Student Grades

* Create a list to store and caculate average grades for students

```python
# organizing student grades
student_grades = [85, 92, 78, 90, 88]

#Adding a new grade
student_grades.append(95)

#Calculating the average grade
average_grade = sum(student_grades) / len(student_grades)
print(f"The average grade is: {average_grade:.2f}")

# Finding the highest and lowest grades
highest_grade = max(student_grades)
lowest_grade = min(student_grades)
print(f"The highest grade is: {highest_grade}")
print(f"The lowest grade is: {lowest_grade}")
```

### Example 3: Managing An Inventory

* Use a list to manage inventory items in a store

```python
# Managing an inventory
inventory = ["Apples", "Bananas", "Oranges"]


# Adding new items to the inventory
inventory.append("Grapes")
inventory.append("Mangoes")

# Removing an item from the inventory
inventory.remove("Bananas")

# Checking if an item is in the inventory
if "Oranges" in inventory:
    print("Oranges are available in the inventory.")

else:
    print("Oranges are not available in the inventory.")


# Print the current inventory
print("Current Inventory:")
for item in inventory:
    print("-", item)
```

### Example 4: Collecting User Feedback

* Use a list to collect and analyze user feedback

```python
# collecting user feedback
user_feedback = ["Greet service", "Good quality", "Could be better", "Excellent experience"]

# Adding new feedback
user_feedback.append("Average experience")

#counting specific feedback
positive_feedback_count = user_feedback.count("Good quality")
print(f"Number of positive feedbacks: {positive_feedback_count}")

# Print all user feedback
print("User Feedback:")
for feedback in user_feedback:
    print("-", feedback)
    
```
