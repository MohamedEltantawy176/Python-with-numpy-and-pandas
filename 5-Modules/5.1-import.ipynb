### Importing Modules in Python: Modules and Packages

In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math 
int(math.sqrt(16))  # Example code to demonstrate the use of math module
```

```python
from math import sqrt, pi
print(int(sqrt(25)))  # Example code to demonstrate the use of sqrt function directly
print(pi)  # Example code to demonstrate the use of pi constant directly
```

```python
import numpy as np # Importing numpy module alias np
array = np.array([1, 2, 3, 4, 5])
print(array)  # Example code to demonstrate the use of numpy array
```

```python
from math import * # Importing all functions from math module
print(factorial(5))  # Example code to demonstrate the use of factorial function directly
```

```python
## using custom package
from package.maths import addition
result = addition(5, 10)
print(result)  # Example code to demonstrate the use of addition function from package  
```
