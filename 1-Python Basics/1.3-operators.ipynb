# Deep Dive into Operators

```python
## Arithemtic Operators in Python
## Arithmetic operators are used to perform mathematical operations such as addition, subtraction, multiplication, and division.
a = 15
b = 4      
addition = a + b          # 19
subtraction = a - b       # 11
multiplication = a * b    # 60
division = a / b          # 3.75
floor_division = a // b   # 3
modulus = a % b           # 3
exponentiation = a ** b   # 50625
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Floor Division:", floor_division)
print("Modulus:", modulus)
print("Exponentiation:", exponentiation) 
```

```python
## Comparison Operators in Python
## Comparison operators are used to compare two values and return a boolean result (True or False).
x = 10
y = 20
equal = (x == y)            # False
not_equal = (x != y)        # True
greater_than = (x > y)      # False 
less_than = (x < y)         # True
greater_equal = (x >= y)    # False
less_equal = (x <= y)       # True
print("Equal:", equal)
print("Not Equal:", not_equal)
print("Greater Than:", greater_than)
print("Less Than:", less_than)
print("Greater Than or Equal:", greater_equal)
print("Less Than or Equal:", less_equal)
```

```python
## Logical Operators in Python
## Logical operators are used to combine conditional statements.
a = True
b = False
and_operator = a and b      # False
or_operator = a or b        # True
not_operator = not a        # False
print("AND Operator:", and_operator)
print("OR Operator:", or_operator)
print("NOT Operator:", not_operator)
```
