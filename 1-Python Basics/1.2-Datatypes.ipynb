# DataTypes

## 1. Definition:
    * Data types are a classification of data which tell the compiler or interpreter how the program intends to use the data.
    * They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.

## 2. Importance of Data Types in Programming 

    Explanation:
    
    * Data types ensure that data is stored in an efficient way.
    * They help in performing correct  operations on data.
    * Proper use of data types can prevent errors and bugs in the program. 

```python
## Integers in Python
## Integers are whole numbers, positive or negative, without decimals, of unlimited length.
a = 10
b = -5
c = 0
```

```python
## Floating-Point Numbers in Python
## Floating-point numbers, or floats, are numbers that contain a decimal point or are in exponential notation.
x = 3.14
y = -0.001
z = 2.5e3  # This is 2.5 times 10 to the power of 3, or 2500.0      
```

```python
## Strings in Python
## Strings are sequences of characters enclosed in single quotes (' '), double quotes (" "), or triple
## quotes (''' ''' or """ """).
str1 = 'Hello, World!'
str2 = "Python is fun."
str3 = '''This is a multi-line
string.'''
str4 = """Another multi-line
string."""
print(str1)
print(str2)
print(str3)
print(str4)
```

```python
## booleans in Python
## Booleans represent one of two values: True or False.
is_active = True
is_logged_in = False
print(is_active)
print(is_logged_in)
```

```python
## common error with datatypes
# Trying to add a string and an integer will raise a TypeError
# Uncommenting the following line will cause an error
# result = "The answer is: " + 42  # This will raise a TypeError
# Correct way to concatenate string and integer is to convert the integer to a string
result = "The answer is: " + str(42)
print(result)
```
