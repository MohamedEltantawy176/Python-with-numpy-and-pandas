### Magic Methods

Magic methods in Python, also konwn as dunder methods (double underscore methods), are special methods that start and end with double underscores. These enable you to define the behavior of objects for built-in operations, such as arthmetic operations, comparisons, and more.


--------------------------------------

Magic methods are predefined methods in Python that you can override to change the behavior of your objects. Some common magic methods include:

* __init__: initialize a new instance of a class.
* __str__: Returns a string representation of an object.
* __repr__: Returns an official string representation of an object.
* __len__: Returns the length of an object.
* __getitem__: Gets an item from container.
* __setitem__: Sets an item in a container.

```python
## Basic magic methods
class Person:
    def __init__(self, name, age):
        self.name = name 
        self.age = age 
    
    def __str__(self):
        return f"{self.name}, {self.age} years old "
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("Mohamed", 32)
print(person)
print(repr(person))
```
