## Decorators

Decorators are a powerful and flexible feature in Python that allows you to modify the behavior of a function or class method. They are commnly used to add functionality to functions or methods without modifying their acrtual code. This lesson covers the basics of decorators, including how to create and use them.

```python
## function copy
def welcome():
    return "Welcome to the advanced python course"

welcome()
```

```python
wel = welcome
print(wel())
del welcome
print(wel())
```

```python
print(wel())
```

```python
##closures
#means function inside a function
def main_welcome():
    msg ="welcome"
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(msg)
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome()
```

```python
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func('Welcome everyone')
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(print)
```

```python
def main_welcome(func, li):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        print(func(li))
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
main_welcome(len,[1,2,3,4,5])
```

```python
## Decorator
def main_welcome(func):
    def sub_welcome_method():
        print("Welcome to the advance python course")
        func()
        print("Please learn these concepts properly")
    return sub_welcome_method()
```

```python
def course_introduction():
    print("This is an advanced python course")

course_introduction()
```

```python
main_welcome(course_introduction)
```

```python
# @ keyword creates the decorator
@main_welcome
def course_introduction():
    print("This is an advanced python course")
```

```python
## example 
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper # not calling the function here 
```

```python
@my_decorator
def say_hello():
    print("hello")
```

```python
say_hello()
```

```python
## Decorators with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def say_hello():
    print("hello")
```

```python
say_hello()
```
