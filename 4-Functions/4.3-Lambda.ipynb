### Lambda Functions in Python

Lambda  functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only one expression. 
They are commonly used for short operations or as arguments to higher-order functions.

```python
# Syntax
lambda arguments: expression

# Example 1: Simple Lambda Function to Add Two Numbers
add = lambda x, y: x + y
result = add(3, 5)
print(result)  # Output: 8
```

```python
def even_or_odd(n):
    return "Even" if n % 2 == 0 else "Odd"
result = even_or_odd(10)
print(result)  # Output: Even

#  Lambda Function to Check if a Number is Even or Odd
even_or_odd = lambda n: "Even" if n % 2 == 0 else "Odd"
result = even_or_odd(7)
print(result)  # Output: Odd
```

```python
## multiple arguments lambda vs regular function

def multiply(x, y):
    return x * y
result = multiply(4, 6)
print(result)  # Output: 24

# Lambda Function to Multiply Two Numbers
multiply = lambda x, y: x * y
result = multiply(3, 7)
print(result)  # Output: 21

```

```python
## map()
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers)) # Using lambda with map to square each number
print(squared)  # Output: [1, 4, 9, 16, 25]
```
