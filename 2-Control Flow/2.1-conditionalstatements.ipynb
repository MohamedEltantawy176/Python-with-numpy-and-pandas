# Conditional Statements (if, elif, else)

```python
## if statements in Python
# An if statement is used to test a specific condition. If the condition evaluates to True, the block of code within the if statement is executed.
age = 18
if age >= 18:
    print("You are eligible to vote.") 
```

```python
## else statements in Python
# An else statement is used to define a block of code that will be executed if the condition in the if statement evaluates to False.
age = 16
if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote yet.")
```

```python
## elif statements in Python
# An elif statement is used to check multiple conditions. If the first condition is False, it
# checks the next condition, and so on.
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
    
    
```

```python
## Nested if statements in Python
# Nested if statements are if statements placed inside another if statement. They allow for more complex decision
# making by checking additional conditions within the scope of the first condition.
num = 10
if num > 0:
    if num % 2 == 0:
        print("The number is positive and even.")
    else:
        print("The number is positive and odd.")
else:
    print("The number is not positive.")
```

```python
## practical examples of conditional statements
# determine if a year is a leap year
year = 2020
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
## Assigment
## Simple Calculator program that performs addition, subtraction, multiplication, and division based on user input with if. elif statements.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")
if operation == '+':
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operation == '-':
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operation == '*':
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation.")
```
