# Loops

```python
## for loop in Python  
# range(5) generates numbers from 0 to 4 
for i in range(5):
    print(i)
```

```python
for j in range(1, 6):
    print(j)
```

```python
for i in range(1, 10, 2):
    print(i)  # Generates numbers from 1 to 9 with a step of 2
```

```python
for i in range(10, 1, -1):
    print(i)  # Generates numbers from 5 to 1 in descending order
```

```python
for j in range(10, 1, -2): 
    print(j) # Generates numbers from 10 to 2 with a step of -2
```

```python
## strings 
str= "Mohamed Eltantawy"

for char in str: 
    print(char)
```

```python
## while loop in Python
count = 0
while count < 5:
    print("Count is:", count)
    count += 1  # Increment count by 1
```

```python
count = 0
while count % 2 == 0:
    print("Even Count is:", count)
    count += 1  # Increment count by 1
```

```python
## Loop Control Statements
# Using break statement
# break statement to exit the loop permanently
for i in range(10):
    if i == 5:
        break  # Exit the loop when i is 5
    print(i)
```

```python
##continue statement
# continue statement to skip the current iteration and move to the next one
for i in range(10):
    if i % 2 == 0:
        continue  # Skip even numbers
    print(i)  # This will print only odd numbers
```

```python
## pass statement
# pass statement as a placeholder
for i in range(5):
    if i < 3:
        pass  # Placeholder, does nothing
    else:
        print(i)
```

```python
## Nested Loops
for i in range(3):  # Outer loop
    for j in range(2):  # Inner loop
        print(f"Outer loop i={i}, Inner loop j={j}") 
```

```python
## Example: Calculate the sum of the first N natural numbers using a while and for loop

## while loop
N = 10
sum_while = 0
count = 1
while count <= N:
    sum_while += count
    count += 1
print("Sum using while loop:", sum_while)

## for loop
sum_for = 0
for i in range(1, N + 1):
    sum_for += i
print("Sum using for loop:", sum_for)
```

```python
## Example: Prime numbers between 1 and 100 using nested loops
for num in range(1, 101): 
    if num >1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```

## Conclusion:
Loops are powerful constructs in Python that allow you to execute a block of code multiple times.
By understanding and using for and while loops, along with loop control statements like break, continue and pass, you can handle a wide range of programming tasks efficiently.
