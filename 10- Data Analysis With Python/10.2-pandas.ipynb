## Pandas-DataFrame And Series

Pandas is a pwerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneoustabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
# A Pandas series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

data = [1,2,3,4,5]
series = pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
## Create a Series from dictionary
data =  {'a': 1,  'b': 2, 'c': 3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]
index = ['a', 'b', 'c']
pd.Series(data, index)
```

```python
## Dataframe
# create a Datafram from a dictionary of a list

data = {
    'Name': ['Mohamed', 'Zain', 'Omar'],
    'Age': [32, 4, 22],
    'City': ['Giessen', 'Marburg', 'Berlin']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a DataFrame from a list of Dictionaries

data =[
      {'Name':'Mohamed', 'Age': 32, 'City': 'Giessen'},
      {'Name':'Zain', 'Age': 4, 'City': 'Marburg'},
      {'Name':'Omar', 'Age': 22, 'City': 'Berlin'},
      {'Name':'Jack', 'Age': 34, 'City': 'Frankfurt'}
]

df=pd.DataFrame(data)
print(df)
```
